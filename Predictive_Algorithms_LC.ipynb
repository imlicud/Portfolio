# Breast Cancer Database

As a first step, we will import a breast cancer database with information about 569 patients with tumors, of which 357 were benign and 212 malignant.

As we can see in the table below, the 30 predictor variables that we will use in the analysis are quantitative (numeric), so no transformation of qualitative variables to numerical values ​​is required. Although we could perform an analysis of each variable and probably do some transformation, correlation analysis, etc., for these algorithm tests we will take the data without modification.

```
import pandas as pd
from sklearn.datasets import load_breast_cancer

data = load_breast_cancer()
pd.DataFrame(data.data, columns=data.feature_names)
```

# Training and Test Data

Before training the algorithms, we need to separate a part of the data that we will use as a test (validation) group. The idea is that we can test our ML classifier on data you've never seen before and see how it performs.

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, random_state=18)
```

# Machine Learning Algorithms

Next we will test the different ML algorithms that we have studied in the course. For this, we are going to train our algorithm on the training data (X_train, y_train) and test it on the test data (X_test, y_test). To find out how our algorithm performed, we will look at the confusion matrix for each algorithm.

## Random Forest

```
#Importamos el algoritmo de sklearn
from sklearn.linear_model import LogisticRegression
#Cargar la forma de calificar (métricas)
from sklearn.metrics import confusion_matrix

#Configuramos el algoritmo
clf=LogisticRegression()
#Entrenar el bosque aleatorio (con los datos de entrenamiento)
clf.fit(X_train, y_train)
#Generar predicciones (predecir los datos de prueba)
clf_predict = clf.predict(X_test)

#Aciertos y Errores

print(confusion_matrix(y_test,clf_predict))

tn, fp, fn, tp = confusion_matrix(y_test,clf_predict).ravel()

print('Predicción Positiva significa tumor benigno, Predicción Negativa significa tumor maligno\n')
print("Verdaderos Positivos: "+str(tp))
print("Falsos Positivos: "+str(fp))
print("Verdaderos Negativos: "+str(tn))
print("Falsos Negativos: "+str(fn))
```

## Neural Network (MLP)

```
from sklearn.neural_network import MLPClassifier

clf=MLPClassifier()
clf.fit(X_train,y_train)
clf_predict = clf.predict(X_test)
tn, fp, fn, tp = confusion_matrix(y_test,clf_predict).ravel()

print('Predicción Positiva significa tumor benigno, Predicción Negativa significa tumor maligno\n')
print("Verdaderos Positivos: "+str(tp))
print("Falsos Positivos: "+str(fp))
print("Verdaderos Negativos: "+str(tn))
print("Falsos Negativos: "+str(fn))
```
